### 4. Data Modeling

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(oov_token="<OOV>") #<OOV> berfungsi sebagai nilai ketika kata pada data testing tidak ada pada data training 
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['content'], df['sentiments'], test_size=0.4, shuffle=True)
```

```python
# proses memasukan data pada database yaitu field content dan sentiments ke dalam variabel array
import numpy as np
training_sentences = []
training_labels = []
testing_sentences = []
testing_labels = []
for row in x_train:
    training_sentences.append(str(row))
for row in y_train:
    training_labels.append(row)
for row in x_test:
    testing_sentences.append(str(row))
for row in y_test:
    testing_labels.append(row)
```

```python
vocab_size = 40000
embedding_dim = 16
max_length = 120
trunc_type = 'post'
oov_tok = '<OOV>'
padding_type = 'post'
```

```python
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(training_sentences)
word_index = tokenizer.word_index


table=PrettyTable(['Kata', 'Urutan'])
x=1
for i in word_index:
    table.add_row([i,x])
    x+=1
# word_index 
print(table)
# word_index
```

```python
sequences = tokenizer.texts_to_sequences(training_sentences)
padded = pad_sequences(sequences, maxlen=max_length, truncating=trunc_type)
testing_sentences = tokenizer.texts_to_sequences(testing_sentences)
testing_padded = pad_sequences(testing_sentences, maxlen=max_length)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim,       input_length=max_length),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
training_labels_final = np.array(training_labels)
testing_labels_final = np.array(testing_labels)
testing_labels_final
```

```python
num_epochs = 30
history = model.fit(padded, training_labels_final, epochs=num_epochs, validation_data=(testing_padded, testing_labels_final))
```

# EVALUASI

<b style="font-size:20px;">VISUALISASI DATA TRAINING DAN TESTING<b>

```python
# VISUALISASI DATA TRAINING DAN TESTING
import numpy as np

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epoch = np.arange(0,len(acc),1)

#proses visualisasi
plt.figure(figsize=(20,9))

# bagian visualisasi history akurasi training dan testing
plt.subplot(1,2,1)
sns.lineplot(x=epoch,y=acc,label='Akurasi Training')
sns.lineplot(x=epoch,y=val_acc,label='Akurasi testing')
plt.xlabel("Epoch")
plt.title("Accuracy Training/Validation")
plt.xticks(epoch,epoch)

# bagian visualisasi history loss training dan testing
plt.subplot(1,2,2)
sns.lineplot(x=epoch,y=loss,label='loss training')
sns.lineplot(x=epoch,y=val_loss,label='loss testing')
plt.xlabel("Epoch")
plt.title("Loss Training/Validation")
plt.xticks(epoch,epoch)
plt.show()
```

<b style="font-size:20px;">TESTING<b>

```python
# pada bagian ini akan mengetes program yang sebelumnya telah dilatih dari data training

#deklarasi variabel predict_sentences yang fungsinya untuk mengkonfersi huruf huruf yang dimasukan menjadi nilai angka.
predict_sentences = tokenizer.texts_to_sequences(["gamenya bagus"])

#deklarasi variabel predict_padded yang berfungsi memasukan angka agnka hasil konfersi dari perkata tadi kedalam matriks
predict_padded = pad_sequences(predict_sentences, maxlen=max_length)

#deklarasi variabel "" untuk menampilkan nilai prediksi dari variabel predict_padded
hm=model.predict(predict_padded)
if  hm<= 0.5:
    print("Jelek")
elif hm > 0.5:
    print("Bagus")

```
